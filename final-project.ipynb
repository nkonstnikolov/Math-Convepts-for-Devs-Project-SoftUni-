## What is a simulation?

A simulation is a model or a replica of a system or an environment used to study and predict the actual behavior of the system or environment by representing its operation over time, without carrying out the experiment on the real system or environment. There are many fields that use simulations, the most common being the technological sphere, which uses simulations for tuning and optimization purposes. Simulations are also used in safety engineering, education, the armed forces, video game development, as well as in most scientific areas, such as weather or astronomy, among others.

When the simulation is carried out by means of a computing device, it is called a digital simulation. Before the dawn of computers, analog simulations have also existed. The commonly used method was to create a decent replica of an environment and scale it down to a working size to observe the effects of the performed manipulations. One example of an analog simulation is the Waterloopbos (Dutch for “waterway forest”) in the Netherlands, which hosted a series of scaled-down models of estuaries and harbors, used in the 1950s to simulate and predict the influence of hydraulic engineering works on the course of water.

## Advantages of doing simulations

Simulations are a great method for analysis and problem solving because they are easily verifiable, communicated and understood. Simulations are efficient because they are able to give us clear insight into the behavior of complex systems. Here are the main advantages of using simulations:

1. Risk-free environment: simulations provide us a safe way to test and explore different potential scenarios. It allows us to make the right decisions before making actual real-world changes.


2. Saving money and time: carrying out tests with simulations is less expensive and less time-consuming than experimenting with real assets.


3. Visualization: simulations can be carried out in 2D as well as 3D, which allows us to not only draw conclusions from numerical data, but also see the real thing in action and demonstrate our findings in a clear and easy way.


4. Accuracy: simulations can be made in such a detailed way that they can provide us with increased accuracy and more precise forecasting than by using traditional analytical models. This can significantly cut costs by optimizing asset usage and knowing our needs in advance.


5. Handling uncertainty: uncertainty in outcome can easily be coded in simulation models, allowing risk quantification and more flexibility when in decision-making and solution-finding.

## Links between statistics and simulation

When a certain hypothesis needs to be tested, it requires some form of data to be processed in order to draw conclusions which will help us to validate or refute the hypothesis. Whenever scientists do not dispose of actual empirical data, an option is to use simulated data. This field of statistics is called Simulation Statistics. It uses artificially generated data for testing purposes. This method has, of course, its own set of advantages and inconvenients, the advantage being that approximating data is a much faster process than actually collecting said data, thus allowing a test to be executed more quickly. However, the major inconvenient is that we are merely dealing with approximations of real-world data, in which a resonable margin of error should be defined, as it obviously would not be a perfect representation of the real-world model that we are trying to describe.

## How can we simulate unknown, random processes?

The simulation of a system with variables that can change randomly with individual probabilities is called a stochastic simulation. The word stochastic comes from the Greek word *στοχαστικός* which means ‘based on guesswork’. The general procedure for carrying out a stochastic simulation goes as follows: we use a method to generate observations of random variables (for example dice throwing), which we then record and repeat the process with a new set of random values, until we gather a sufficient amount of data. In the end, the distribution shows the most probable estimates, as well as the range(s) of values the variables are most likely to fall in. Here is a quick example of a basic simulation statistics model based on dice throwing:


<u>Step 1: Define outcomes</u>

Let's assume that we are throwing a 6-sided die. In this case, we can get any one of six outcomes :

### $A = \{ 1,2,3,4,5,6 \}$

<u>Step 2: Calculate probability of desired outcomes</u>

Based on our question, the desired outcome is rolling one particular of the six numbers (for the sake of example, let's say the number 6).

```python
def calculate_probability(desired_outcomes, possible_outcomes):
    probability = str(desired_outcomes / possible_outcomes)
    print("The theoretical probability of rolling a 6 is: " + probability + ".")
    
calculate_probability(1, 6)
```

Alright, so our theoretical probability to roll the number 6 turns out to be 0.1(6). Let's move forward.

<u>Step 3: Generate random values</u>

Now we're getting into the real simulation part. Using a random-number generator, let's generate a dataset of 1000 numbers going from 1 to 6.



```python
values_array = []
for i in range(1000):
    value = randint(1, 6)
    values_array.append(value)
print(values_array)
```

<u>Step 4: Analyze data</u>

Now that we have chosen a number, and we know that its theoretical probability to be rolled is 0,1(6), let's see what the simulated data shows us: let us calculate the empirical probability.

```python
chosen_number = np.array(values_array)
a = np.count_nonzero(chosen_number == 6)
all_numbers = len(values_array)
empirical_probability = str(a / all_numbers)
print("The empirical probability of rolling a 6 in this dataset is: " + empirical_probability + ".")
```

Out of curiosity, let's see how the values look like in a histogram:

```python
plt.hist(values_array,bins=[1,2,3,4,5,6,7], ec = "black")
plt.xlabel("number")
plt.ylabel("times rolled")
plt.show()
```

As we can see from the histogram above, the empirical probability is always bound to differ from the theoretical probability.

## Monte Carlo Methods

The Monte Carlo methods are a type of computational algorithms that we use to obtain numerical results from random sampling techniques. We use the Monte Carlo methods to make estimations. If we need to know the average value of a random variable, but its distribution is not determined, and if it is possible to take samples from the distribution, we can estimate it by finding the average of all samples. If we have enough samples, then according to the law of large numbers, their average value should be close to the theoretical value of the variable. According to the central limit theorem, this average will have a Gaussian distribution around the real value.

We can apply the Monte Carlo method to simulating events for which we want to find out what will happen over a long period of time. By generating a series of random values, we can apply those values to the event model we are simulating, we can run the simulation a great number of times (the bigger the better) and then collect and analyze the results, and draw conclusions.

## Numerical Integration Using Monte Carlo Methods

In analysis, numerical integration is used for calculating the value of a function's definite integral. The basic problem in numerical integration is to compute an approximate solution to a definite integral $\int_{a}^{b}f(x)dx$. There are several methods to do this. One of them is the trapezoidal rule. Another is Monte Carlo integration. Let's explore Monte Carlo integration and, later, see how it compares with the trapezoidal rule integration in terms of performance.

Monte Carlo integration is a technique that uses random numbers. Unlike other algorithms that evaluate the integrand at a regular grid, Monte Carlo integration uses an alternative method of randomly choosing points at which the integrand is evaluated.

Here's how Monte Carlo integration basically works:

We can approximate the integral $\int_{a}^{b}f(x)dx$ by generating a number of random values that fall within the interval from point **a** to point **b**:

### $\int_{a}^{b}f(x)dx \approx (b-a)\frac{1}{N}\sum_{i}f(x_{i})$

^ where $(b-a)$ is the distance between **b** and **a** and $\frac{1}{N}\sum_{i}f(x_{i})$ is the expanded definition of the average of a function (1 / the number of points used times the sum of the function evaluated at these points).

Let's try to implement this for the integral of $sin(x)$ from 0 to $\pi$ with 1000 iterations:

```python
start_time = time.time()

a = 0 # lower limit
b = np.pi # upper limit
N = 1000 # number of iterations
x = random.uniform(a,b,N) # this generates N numbers between the limits we've set

def function(x):
    return np.sin(x)

integral = 0.0

for i in range(N):
    integral += function(x[i])
    
answer = (b-a)/float(N)*integral

print("The integral from 0 to π of sin(x) is:", answer)

print("This operation took %s seconds to execute" % round((time.time() - start_time), 10))
```

As we can see, we're already quite close to 2, which is the real value of this integral. Now let's see how this looks like in a histogram:

```python
areas = []

for i in range(N):
    x = np.zeros(N)

    for i in range(len(x)):
        x[i] = random.uniform(a,b)
        integral = 0.0

    for i in range(N):
        integral += function(x[i])

    answer = (b-a)/float(N)*integral
    areas.append(answer)

plt.hist(areas, bins = 30, ec = 'black')
plt.xlabel("areas")
plt.show()
```

As we can see, the highest values are centered around 2. Let's see if we can get even closer if we increase the number of iterations:

```python
a = 0 # first limit
b = np.pi # second limit
N = 100000 # number of iterations
x = random.uniform(a,b,N) # this generates N numbers between the limits we've set

def function(x):
    return np.sin(x)

integral = 0.0

for i in range(N):
    integral += function(x[i])
    
answer = (b-a)/float(N)*integral

print("The integral from 0 to π of sin(x) is:", answer)
```

As we can see, when we increased the number of iterations a hundred-fold, we got significantly closer to the real value of 2.

Let's now try another function, namely $\int_{0}^{2\pi}cos^2(\theta)d\theta$, whose value is $\pi$ and let's increase the iterations to a million for the sake of accuracy.

```python
a = 0
b = np.pi * 2
N = 1000000
theta = random.uniform(a,b,N)

def function(theta):
    return np.cos(theta) * np.cos(theta)

integral = 0.0

for i in range(N):
    integral += function(theta[i])
    
answer = (b-a)/float(N)*integral

print("The integral from 0 to 2π of cos²(θ) is:", answer)
```

As we can see, the accuracy of this is close enough to $\pi$ that we manage to get 3.14 every time we run this.

I included a time counter in the first operation for a reason. Now we'll see how the Monte Carlo integration method compares time-wise, and also accuracy-wise with another integration method, namely the Trapezoidal rule technique. Let's calculate the integral from 0 to $\pi$ of $sin(x)$ using this technique and compare it to our earlier results. For comparison reasons, we'll use 1000 trapezoids, just like we used 1000 iterations in the previous example, and we'll also measure the time of execution.

We'll write a function that calculates the sum of the areas of the trapezoids using the following formula:

### $I = h\left \{ \frac{1}{2}[f(x_a)+f(x_b)]+f(x_1)+f(x_2)+...+f(x_{n-2})+f(x_{n-1}) \right \}$

```python
start_time = time.time()

f = lambda x: np.sin(x) # defining the function
a = 0 # lower limit
b = np.pi # upper limit
n = 1000 # number of trapezoids
h = (b - a) / n # step size
S = 0.5 * (f(a) + f(b))
for i in range (1, n):
    S += f(a + i * h)
answer = h * S
print ("The integral from 0 to π of sin(x) is:", answer)
print("This operation took %s seconds to execute" % round((time.time() - start_time), 10))
```

### Result and comparison:

The Monte Carlo method calculated the integral from 0 to $\pi$ of $sin(x)$ as **2.0117109007303458**. The operation took **0.0019981861** seconds to execute.

The trapezoidal rule method calculated it as **1.9999983550656624**, significantly closer to the real value. The operation took **0.0019984245**, which is just 0.0000002384 seconds slower than the Monte Carlo method. This time difference is so small, it is completely insignitifcant.

Conclusion: The trapezoidal rule method calculates a better approximation of integrals, at the expense of it being slower, but barely enough to be noticeable, especially for a smaller number of iterations.
